# Fato chamada

Este notebook carrega o fato chamada

```
schema = spark.read.parquet('/dados/chamadas').schema
chamadas = spark.readStream.schema(schema).parquet('/dados/chamadas')
```

```
chamadas.writeStream.format('delta').option("checkpointLocation",'/_chk/fat_chamada').option("mergeSchema", "true").start('/delta/fat_chamada')
```

```
%sql

CREATE TABLE IF NOT EXISTS fat_chamada
USING delta 
OPTIONS
(
  path = "/delta/fat_chamada"
)
```

```
%sql select * from fat_chamada
```
