Simula a geração de chamadas de um call center

Gera as colunas de operador e sucesso da chamada

```
import pyspark.sql.functions as F

chamadas = spark.readStream.format('rate').option("rowsPerSecond", 1).load()


chamadas = chamadas.withColumn('id_operador',(F.col('value') % 3)+1)\
                   .withColumn('in_sucesso',(F.rand()>0.7))\
                   .withColumn('in_telefone',F.lit('sim'))
```

Grava os dados em arquivos parquet

```
chamadas.writeStream.format('parquet')\
                    .option("checkpointLocation",'/_chk/chamadas')\
                    .start('/dados/chamadas')
```
